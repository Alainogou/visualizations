```python
import pandas as pd
import matplotlib.pyplot as plt


x=[1, 4, 5, 6, 7]
y=[2, 6, 3, 6, 3]


plt.plot(x, y, 'r--', x, y , 'bo' )
# Titre et labels
plt.title('Display marker')
plt.xlabel('x - axis')
plt.ylabel('y - axis')

# Limites des axes
plt.xlim(1, 8)
plt.ylim(1, 8)

# Affichage du graphique
plt.show()
```
