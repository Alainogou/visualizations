```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Define the number of rows and columns for the subplots
nrows = 2
ncols = 3

# Create the figure and axes with spacing
fig, axes = plt.subplots(nrows, ncols, figsize=(10, 6), gridspec_kw={'hspace': 0.5, 'wspace': 0.5})

# Loop to create each subplot
for i in range(nrows * ncols):
    # Calculate the position of the subplot (row and column)
    row = i // ncols
    col = i % ncols

    # Access the current subplot
    ax = axes[row, col]

    # Add text to the center of the subplot
    ax.text(0.5, 0.5, f'(2,3,{i + 1})',
            ha='center', va='center', fontsize=12, fontweight='bold')

    # Add a title to the subplot
    ax.set_title(f'Title {i + 1}')

   

# Display the graph
plt.show()
```
