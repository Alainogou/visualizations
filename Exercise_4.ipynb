```python
import pandas as pd
import matplotlib.pyplot as plt

import pandas as pd
import matplotlib.pyplot as plt

left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]

fig, y_left = plt.subplots()

y_right = y_left.twinx()

y_left.plot( x_axis, left_data, color='black' )
y_right.plot( x_axis, right_data, 'r')

y_left.set_ylabel('Big')
y_right.set_ylabel('Small')
fig.suptitle('Twin axes')

plt.show()
```
